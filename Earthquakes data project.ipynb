## Activity 8: Dictionaries!

Create a dictionary of how many earthquakes happen in each country. The country field is named **COUNTRY**.

- Break this up into two cells.
    - In the first cell, build the complete dictionary from an empty one. This took 6 lines of code using a for loop.
    - In the second cell, report the number of unique countries in this dictionary. As a sanity check, I had 134.
    
Tips:

- My for loop was "for quake in quakes.COUNTRY:".

```python
count = 0
thisdict = {} # Empty dictionary to add values to 

for quake in quakes.COUNTRY:
    thisdict[count] = quake
    count+=1

print(thisdict) # dictionary with listing on countries with count number
```

```python
uniqueValues = set(thisdict.values())

for value in uniqueValues:
    print(value)
    
print(len(uniqueValues)) # length at bottom returns 134 total unique countries
```

## Activity 9: Dictionaries!

Using the dictionary that you created in the last step, print out a list of all of the countries with 50 or more earthquakes. As a sanity check, I found 17 countries with at least 50 earthquakes in the last 118 years. This took 3 lines of code. This should print both the name of the country and the number of earthquakes in that country. For example, one line was "JAPAN 206".

```python
from collections import Counter

a = Counter(thisdict.values())

for key, value in a.items():
    if value >= 50:
        print(key,value) # print frequencies of countries with greater than or equal to 50 occurences in the dictionary listing
```
